<h2>Histograms</h2>

<h3>A Quick Intro</h3>
<p>
Plot the frequency of the values of your data. So if you have your data as {1,1,1,3,3,4}. It's histogram would look as follow
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
x = np.array([1,1,1,3,3,4])

plt.hist(x, bins = 10)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.show()
```

<h3>How can histograms be used in Data Science</h3>
<p>You can get an accurate measure of probability distribution of your data. Also you can identify the outliers in your data as explained in <a href = "https://www.youtube.com/watch?v=sr80Ag0LD4s&t=2s">this video</a></p>

<h3>What parameters to tune to get a good histogram</h3>
<p>The only parameter that you want to get correct are the number of bins you are using to plot the histogram.</p>
<p>

<h3>What are bins?</h3>
<p>For the given example out min value was 1 and the max value was 4. So if you choose say k number of bins, you would be dividing the region between min and max value into k parts.<p>

```python
fig = plt.figure(figsize = (10,5))
plt.subplot(121)
plt.hist(x, bins = 1)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 1')

plt.subplot(122)
plt.hist(x, bins = 4)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 4')

plt.show()
```

<h3>How to choose the number of bins?</h3>
<p>Bins that are two narrow can cause a lot of noise and hide the important information about the distribution while bins that are two wide can hide the important detials of the distribution.</p>

```python
data = 10*np.random.randn(200,1)
N = 100
```

<h3>1) Square Method</h3>
<img src = "Extras/square.png"></img>

```python
bins = np.sqrt(N)

plt.hist(data)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 10')
plt.show()
```

<h3>2) Sturges formula</h3>
<img src = "Extras/sturges.png"></img>

```python
bins = np.ceil(np.log(N)/np.log(2)) + 1

plt.hist(data)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 10')
plt.show()
```

<h3>3) Rice formula</h3>
<img src = "Extras/rice.png"></img>

```python
bins = 2*np.power(N, 1.0/3)

plt.hist(data)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 10')
plt.show()
```

<h3>4) Scott formula</h3>
<img src = "Extras/scott.png"></img>

```python
bins = (np.max(data, axis = 0) - np.min(data, axis = 0)) / (3.5 * np.std(data, axis = 0) / np.power(N, 1.0/3))

plt.hist(data)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 10')
plt.show()
```

<h3>5) Freedman-Diaconis formula</h3>
<img src = "Extras/freedman.png"></img>

```python
bins = (np.max(data, axis = 0) - np.min(data, axis = 0)) / (2 * stats.iqr(data, axis = 0) / np.power(N, 1.0/3))

plt.hist(data)
plt.xlabel('x values')
plt.ylabel('Frequency of x values')
plt.title('Number of bins = 10')
plt.show()
```
